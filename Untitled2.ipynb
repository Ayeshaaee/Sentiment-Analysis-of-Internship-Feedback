```python
import pandas as pd
df = pd.read_csv('/content/archive (18).zip')
df = df.dropna(subset=['Sentiment'])  # Drop rows with missing feedback
df.head()
```

```python
!pip install -q transformers
from transformers import pipeline
```

```python
# Load sentiment pipeline
sentiment_pipeline = pipeline("sentiment-analysis")

# Get BERT sentiment labels
results = sentiment_pipeline(df['Text'].tolist())
df['Sentiment_Label'] = [res['label'] for res in results]
df['Sentiment_Score'] = [res['score'] for res in results]
df['Sentiment'] = df['Sentiment_Label'].apply(lambda x: 1 if x == 'POSITIVE' else 0)

# Preview
df[['Text', 'Sentiment_Label', 'Sentiment']].head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score

# Split data
X_train, X_test, y_train, y_test = train_test_split(df['Text'], df['Sentiment'], test_size=0.2, random_state=42)

# Vectorization
tfidf = TfidfVectorizer(stop_words='english', max_features=1000)
X_train_tfidf = tfidf.fit_transform(X_train)
X_test_tfidf = tfidf.transform(X_test)

# Train model
lr = LogisticRegression()
lr.fit(X_train_tfidf, y_train)

# Predict
y_pred = lr.predict(X_test_tfidf)

# Evaluate
print("âœ… Logistic Regression Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Bar plot of BERT sentiments
sns.countplot(x='Sentiment_Label', data=df, palette='Set2')
plt.title("Sentiment Distribution (BERT)")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Extract only negative feedback
negative_feedback = " ".join(df[df['Sentiment_Label'] == 'NEGATIVE']['Text'].dropna())

# Generate Word Cloud
wordcloud = WordCloud(width=1000, height=500, background_color='white', colormap='Reds').generate(negative_feedback)

# Display
plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Common Words in Negative Feedback")
plt.show()
```
